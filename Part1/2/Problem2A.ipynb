```python
import numpy as np
import pandas as pd

from IPython.display import HTML, display
from sklearn.cluster import KMeans

import tabulate

train = pd.read_csv('data/train')

train.head()
```

```python
k3 = KMeans(n_clusters=3).fit(train)

display(HTML(tabulate.tabulate(k3.cluster_centers_, tablefmt='html')))
```

```python
k5 = KMeans(n_clusters=5).fit(train)

display(HTML(tabulate.tabulate(k5.cluster_centers_, tablefmt='html')))
```

```python
k10 = KMeans(n_clusters=10).fit(train)

display(HTML(tabulate.tabulate(k10.cluster_centers_, tablefmt='html')))
```
